# Missão 1: Time de vendas

Precisamos construir uma visualização para indicar os **10 maiores clientes por total de venda**.

Com isso, a equipe de vendas pode pensar em estratégias para manter esses clientes comprando em nossa loja.

### Passo 01: Agrupando os dados dos 10 maiores clientes

```python
df_clientes_vendas = df.groupby(['nome_cliente'])['vendas'].sum().nlargest(10).copy()
df_clientes_vendas = df_clientes_vendas.reset_index()
df_clientes_vendas
#O nlargest faz com que os 10 maiores clientes sejam rankeados
```

```python
# Renomeando as colunas
df_clientes_vendas.columns=['Clientes', 'Vendas']
df_clientes_vendas
```

```python
#Redefinindo os valores do index
df_clientes_vendas['Ranking'] = df_clientes_vendas.index+1
df_clientes_vendas.set_index('Ranking', inplace = True)
df_clientes_vendas
```

```python
# Transformando o DF em um OBJETO STYLER
s = df_clientes_vendas.style
s
#Nota: agora n é mais um DF, é um objeto!!!
```

```python
#Ajustando a coluna vendas (deixar os valores certinhos!)
#precisamos colocar a coluna como um dicionario
s.format({'Vendas': 'R${:,.2f}'})
```

# Missão 2: Time comercial

Precisamos construir uma visualização que permita associar o **total ganho em vendas e o total lucrado por tipo de produto**.

Isso vai permitir que o time comercial possa focar em estratégias para alavancar as rendas que ganham com cada tipo de produto vendido na loja.

```python
df_venda_lucro = df.groupby(['tipo_produto'])[['vendas', 'lucro']].sum()
df_venda_lucro.index.name = 'Tipo de Produto'
df_venda_lucro
```

```python
#transformando o Df em um objeto
estilo_produto = df_venda_lucro.style
estilo_produto
```

```python
#estilizando a tabela
  # ajustando as unidades
  # marcando quais sao os valores max e min dentro do objeto (highlight)  -> Podemos salvar em html (ex: #F16165) ou em formato texto(ex: 'ligthgreen)
estilo_produto.format('R$ {:,.2f}')\
              .highlight_max(color = 'darkgreen')\
              .highlight_min(color = '#F16165')
estilo_produto
```

```python
#Criando um gradiente (Fica melhor que antes)
estilo_produto.format('R$ {:,.2f}')\
              .background_gradient(cmap = 'Greens')
estilo_produto
```

```python
#Ajustar a primeira coluna e linha para ficar mais fluido com o grafico
  # por meio do CSS
    #chaves -> define o tipo de estilização
    #elementos -> a estilização em si
cabecalho = {
   'selector': 'th',
   'props':'font-weight:bold; font-family: Arial; text-align:center; text-transform: capitalize;'
}
# primeiro elemeto = selector: vai seecionar parte da tabela que será transformada
  #th: vai selecionar todo o cabecalho da tabela (primeira linha e os indexes 'primeira coluna')
# segundo elemento = props: sao as propriedades

# para separar as propriedades no css precisamos utilizar o ;

# --> Aplicando a estilização CSS no nosso objeto (visualização)
estilo_produto.set_table_styles([cabecalho], overwrite=False )
estilo_produto
  #overwrite = False -> faz com que nao percamos a transformação do gradiente, demais ajustes na tabela n seja perdido
```

# Missão 3: Time de logística

Precisamos construir uma visualização que permita entender a **distribuição de pedidos por região do Brasil**.

Isso vai permitir que o time logística entenda onde é mais interessante focar material e pessoas.

```python
#Puxando os dados do DF
df_regiao = pd.DataFrame(df['regiao'].value_counts())
df_regiao.columns = ['N pedidos']
df_regiao
```

```python
#alterando o nome da primeira coluna
df_regiao.index.name = 'Região'
df_regiao
```

```python
#calculando o percentual de pedidos
import numpy as np
porcentagem = df_regiao['N pedidos'].to_numpy()
porcentagem = 100*porcentagem/porcentagem.sum()
porcentagem
```

```python
#criando a coluna para colcar as porcentagens
df_regiao['Porcentagem Região'] = porcentagem
df_regiao
```

```python
# Configurando o estilo da tabela
#comecando a estilização
estilo_regiao = df_regiao.style
estilo_regiao
```

```python
#Estilo para o cabeçalho (selector tem que ser o th)
cabecalho = {
    'selector':'th',
    'props':'font-weight: bold; font-family: Arial; text-align: center; background-color: white'
}

#Estilo para as células(selector tem que ser o td)
celulas = {
    'selector':'td',
    'props': 'background-color: white;'
}

estilo_regiao.set_table_styles([cabecalho, celulas])
```

```python
#ajustandoos valores da distribuição
estilo_regiao.format({'Porcentagem Região': '{:.2f}%'})\
              .bar(subset = 'Porcentagem Região', vmin = 0, vmax = 100.0, color = 'lightblue')
```

# Missão 4: Time de logística

Novamente o time de logística deseja repensar sua estrutura e organização de fornecimento de produtos e para isso, foi pensada em uma visualização que permita identificar padrão de **quantidade de produtos solicitados por mês por departamento**.

```python
#Criando um novo DF
df_meses = df.copy()

#ordenar os pedidos de acordo com os meses
df_meses=df_meses.sort_values('data_pedido')
df_meses.head()
```

```python
#Criando uma coluna com ano e mes
df_meses['meses'] = df_meses['data_pedido'].dt.strftime('%Y-%b')
#esse ultimo comando serve para transformar os dados da coluna, que sao e date time, para criar uma formatação que queremos, ela cria uma copia
  # - %b -> vai pegar apenas as tres primeiras letras

#dando uma geral nos index
df_meses = df_meses.reset_index(drop = True)
df_meses.head()
```

```python
# Trabalhando com pivot_table (tabela dinamica)
  #variavel1: departamentos (linhas)
  #variavel2 = meses (colunas)
  #variavel 3 = qtd de produtos somados(relação entre as variaveis 1 e 2)

#criando um novo DF
vendas_mensais = df_meses.pivot_table(index = 'departamento', columns = 'meses', values = 'quantidade', aggfunc = 'sum', sort = False)
vendas_mensais
```

```python
# Visualização fixa (da primeira coluna -> para esse exemplo)

#criando o objeto style
estilo_mensal = vendas_mensais.style
estilo_mensal
  #OBS: quando trazemos o objeto style, ele apresenta TODOS os resultados, diferente do apresentado acima

#fixando a coluna
estilo_mensal.set_sticky(axis = 'index')
  # ao arrastar a coluna para o lado vemos que está fixa! Entretanto, n tem fundo. Isso torna o texto sobreposto :('
    # Isso pode ser melhorado!!! no proximo código!!!
```

```python
#Para facilitar a visualização

#ajustando o cabecalho -> utilizando o .col_heading!!!
colunas = {
    'selector': '.col_heading',
    'props': 'font-weight: normal; font-family: Arial'
}

#ajustando o restante da tabela
tabelas = {
    'selector':'td, th', #pega tanto o cabecalho qt linhas
    'props': 'text-align: left;'
}

#ajustando o index (Ajustando apenas essa coluna que foi indexada, travada)
index = {
    'selector': '.index_name',
    'props': 'font-weight: bold; font-family: Arial; text-align: right;'
}

#
```

```python
estilo_mensal.set_table_styles([colunas, tabelas, index], overwrite = False)
# ao utilizar o overwrite como False o codigo mantem a condição da coluna congelada. Ou seja, n cancela o que foi feito anteriormente
```

# Missão 5: Relatório de performance

O relatório de performance é um documento muito importante para a empresa. Nele, todos que trabalham na loja podem entender como anda a empresa. Por conta disso, tabelas bem construídas são essenciais para mostrar os dados da melhor forma.

Portanto, precisamos construir uma visualização que permita entender a **relação entre tipos de clientes e modo de envio de produtos de acordo com as vendas**. Com isso, será possível destacar no relatório qual o modo de envio e quais tipos de modo de envio mais geraram ganho em vendas.

```python
#criando um novo dataframe

df_cliente = df.pivot_table(index = 'segmento_cliente', columns = 'modo_envio', values = 'vendas', aggfunc = 'sum')
df_cliente
```

```python
# necessario acrescentar o total de vendas por segmento
df_cliente['Total'] = df_cliente.sum(axis=1)
df_cliente.loc['Total'] = df_cliente.sum()
df_cliente
```

```python
#Criando o objeto style
compra_cliente =df_cliente.style.format('{:,.2f}')
compra_cliente
```

```python
#deixando uma visualização mais limpa da tabela

tabela = {
    'selector':'td, th:not(.index_name)',
    'props': 'font-weight: normal; fonte-family: Arial; text-align: center; background-color: white;'
}

index = {
    'selector': '.index_name',         # aqui vai selecionar APENAS o index name!!!
    'props': 'font-weight: normal; text-align: right; font-style: italic; color: black'
}
#podemos deixar o index main mais apagado (neste caso sao: modo_envio e segmento_cliente) para isso:
  #preciso selecionar o selector th:not(.index_main)


#Aplicando o estilo na tabela
compra_cliente.set_table_styles([tabela, index])
```

```python
# Fazendo uma adicao de linhas entre os dados
#código para acrescentar linha de separacao do cabecalho com os elementos abaixo:
compra_cliente.set_table_styles({
    'Total': [{
        'selector':'th',
        'props': 'border-top: 1px solid #181818'
    },
    {
        'selector':'td',
        'props': 'border-top: 1px solid #181818'
    }],
    'B2B':[{
        'selector':'th',
        'props': 'border-top: 1px solid #181818'
    },
    {
        'selector':'td',
        'props': 'border-top: 1px solid #181818'
    }]
}, overwrite = False, axis =1)


```

```python
#Destacando elementos
compra_cliente.set_table_styles({
    'Total': [{
        'selector': '.true',
        'props':'background-color: #d3d3d3'
    }]
},overwrite = False, axis = 0)

#aqui vamos selecionar o elemento da ultima linha, segunda coluna. Isso aqui vai se ligar com o selector acima
cores_coluna = pd.DataFrame(['false', 'true','false'], index=df_cliente['Total'].index,
                            columns=['Total'])

#aplicando a selecao dentro da vissualizacao
compra_cliente.set_td_classes(cores_coluna)
```
