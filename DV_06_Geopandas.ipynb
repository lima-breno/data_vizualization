#Carregando dados dos imoveis do RJ mun


```python
#Transformar dados em geodataframes

import pandas as pd

#Primeiro preciso carregar os dados
url1 = 'https://github.com/alura-cursos/curso_geopandas/raw/main/dados/dados.csv'

df = pd.read_csv(url1,sep ='\t')
df.head()
```

```python
#Transformando a localizacao dos dados em um geodataframe
list(zip(df['Longitude'],df['Latitude']))
```

```python
#passando esses dados em geometry (transformando dados da lat+long em um único ponto)
## Para isso precisamos utiliar a biblioteca shapely

from shapely import Point

# Criando os pontos
pontos_geometricos = [Point(x) for x in zip(df['Longitude'],df['Latitude'])]
pontos_geometricos
```

```python
#Criando um geodataframe com essas infos
  # crs é o sistema de coordenadas geométricas

crs = {'proj': 'latlong', 'ellps': 'WGS84', 'datum': 'WGS84', 'no_defs': True}

geo_dados = gpd.GeoDataFrame(df,crs= crs, geometry = pontos_geometricos )
```

```python
#Criou uma ultima columa com os pontos

#Salvando os dados

geo_dados.to_file('DATASET.shp')
```

```python
geo_dados.plot(alpha = 0.2)
```

```python
# Foram plotados os dados em um grafico
 ## para plotar os dados em um mapa é preeciso que os dados do mapa e dos imoveis tenham o mesmo formato crs

### -> Colocando os dados dentro do mapa

# NEcessario verificar os sistemas de coordenadas
rj.crs
```

```python
geo_dados.crs
```

```python
# As informações estão separadas!!

#Parâmetros do crs:
  #proj: projecao -> cilindrica, azimutal ou plana,conica.
    ##pode ser latlong, utm ou tmerc -> sao tipos de projecoes cilindricas mais usadas
  #ellps: elipsoide:
    #pode ser WGS84, GRS80, CRLK80 -> tipos de elipsoide mais usados
  #datum: auxilia a determinar os eixos e posição central do elipsoide
    # pode ser WGS84 (utilizado pelos sistemas de GPS), SIRGAS (oficialmente utilizado para o Brasil)

  # Projecao UTM - Universal Transverse Mercator (UTM) é uma projeção que apresenta menos distorcao,
    ## como estamos lidando com dados de um pequeno espaço, vai facilitar mais
      ## Ela divide o mapa em 60 partes, precisamos especificar quais as zonas em que estamos trabalhando
      ### O brasil vai da zona 18 a 25 (eixo x)
```

```python
#Modificando o sistema de coordenadas
#Estabelendo as coordenadas

  # qual zona utm o RJ mun faz parte?
    #pesquisar no Chat GPT -> zona 23S
  # Qual o EPSG para a zona UTM 23S utilizando o SIRGAS2000?
    #Pesquisar no chat GPT -> 31983

# Estas infos do EPSG podem ser encontradas com maior precisão no site spatialreference.org
  # aqui deve-se pesquisar pela zona e clicar no SIRGA2000, depois clicar no Proj4
    #esse caminho vai apresentar certinho o crs no geopandas
      #rsultado = +proj=utm +zone=23 +south +ellps=GRS80 +units=m +no_defs +type=crs

```

```python

 # Ajustando os dados crs do mapa
 #podemos passar a unidade para km (para melhorar visualização dos numeros)
rj = rj.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +units=km +no_defs +type=crs')
rj
```

### SErá que o imóvel fica mais caro ou mais barato próximo a estaao do metro?

```python
url_m = 'https://github.com/alura-cursos/curso_geopandas/raw/main/dados/metro.geojson'
#Arquivo em GeoJSON
```

```python
metro = gpd.read_file(url_m)
```

```python
metro.head()
```

```python
# Precisamos ajustar o sistema de referencias para ficar igual ao nosso sistemad e referencias

metro = metro.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +units=km +no_defs +type=crs')
```

```python
#Plotando os dados doo metro na base

base = rj.plot(color = 'white', edgecolor = 'black')
geo_dados.plot(ax = base, alpha= 0.2)
metro.plot(ax=base, color='red', markersize=5)
```

```python
# Verificando se há diferenca no preco do imovel ao considerar a distancia do metro
```

```python
geo_dados.iloc[0].geometry
```

```python
#determinando as distancias dos metros quando comparado com o primeiro ponto
distancias = metro.distance(geo_dados.iloc[0].geometry)
```

```python
#Calculando a menor distancia do metrô para o primeiro imovel
distancias.min()
```

```python
# Calcular a menor dist para cada um dos imoveis
geo_dados['Dist_metro']= geo_dados['geometry'].apply(lambda x: metro.distance(x).min())
```

```python
#VErificando se a correlaçao existe
import numpy as np
geo_dados_num = geo_dados.select_dtypes(include=[np.number])
correlacao = geo_dados_num.corr()
```

```python
#Criando o mapa de calor para verificar a correlacao entre preco x dist

import seaborn as sns

sns.heatmap(correlacao, annot = True, cmap = 'RdYlBu', linewidth = 0.5)
```

### Os demais meios de transporte influenciam o valor do imovel?

```python
# O metro n pega toda a area do munc do RJ. Para melhorar a correlacao podemos
  #acrescentar outras infos no mapa (Dados do transporte pub)
```

```python
#importando os dados do trem e brt

url_b = 'https://github.com/alura-cursos/curso_geopandas/raw/main/dados/brt.geojson'
url_t = 'https://github.com/alura-cursos/curso_geopandas/raw/main/dados/trem.geojson'
```

```python
trem =gpd.read_file(url_t)
brt = gpd.read_file(url_b)
```

```python
trem = trem.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +units=km +no_defs +type=crs')
brt = brt.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +units=km +no_defs +type=crs')
```

```python
#Deixando os dados estao dentro do munc RJ
trem = trem[trem.within(rj.iloc[0].geometry)]
brt = brt[brt.within(rj.iloc[0].geometry)]
```

```python
base= rj.plot(color = 'white', edgecolor = 'black')
geo_dados.plot(ax = base, alpha= 0.2)
trem.plot(ax=base, color='red', markersize=5)
brt.plot(ax=base, color = 'green', markersize = 5)
metro.plot(ax=base, color='black', markersize=5)
```

```python
#Fazendo a correlaçao
transporte = pd.concat([trem.geometry,brt.geometry, metro.geometry], ignore_index = True)
```

```python
geo_dados['Dist_transporte'] = geo_dados['geometry'].apply(lambda x: transporte.distance(x).min())
```

```python
# Fazendo a correlacao

geo_dados_num = geo_dados.select_dtypes(include=[np.number])
correlacao = geo_dados_num.corr()

#Plotando-a

import seaborn as sns

sns.heatmap(correlacao, annot = True, cmap = 'RdYlBu', linewidth = 0.5)
```
